```python
import pandas as pd
import seaborn as sns
```

```python
df_male=pd.read_csv('df_male.csv')
df_female=pd.read_csv('df_female.csv')
char_date=pd.read_csv('char_date.csv')
char_date
```

```python
df_male_date=df_male.merge(char_date, left_on="Unnamed: 0", right_on='character')
df_female_date=df_female.merge(char_date, left_on="Unnamed: 0", right_on='character')
```

```python
df_male_date = df_male_date[['character','text','date']]
df_female_date = df_female_date[['character','text','date']]
```

```python
from tqdm.notebook import tqdm
import random
import re
import numpy as np
def get_words_yearly(words,df_male,df_female,year_min,year_max):
    whole_count=[]
    word_list=[]
    year_list=[]
    for year in tqdm(range(year_min,year_max)):
        word_year_count=[]
        for word in words:
            df_male_temp = df_male[df_male['date']==year]
            df_female_temp = df_female[df_female['date']==year]
            min_words=min(len(df_male_temp),len(df_female_temp))
            if len(df_male_temp)>min_words:
                rd_idx=random.sample(range(0, len(df_male_temp)), min_words)
                df_male_temp_restrict=df_male_temp.iloc[rd_idx]
                cnt_diff=0
                for idx in range(min_words):
                    input_string_male=df_male_temp_restrict['text'].iloc[idx]
                    input_string_female=df_female_temp['text'].iloc[idx]
                    count_word_male = sum(1 for _ in re.finditer(r'\b%s\b' % re.escape(word), input_string_male.lower()))
                    count_word_female= sum(1 for _ in re.finditer(r'\b%s\b' % re.escape(word), input_string_female.lower()))
                    count_word_diff = count_word_female-count_word_male
                    cnt_diff+=count_word_diff
                word_year_count.append(cnt_diff)
            else:
                rd_idx=random.sample(range(0, len(df_female_temp)), min_words)
                df_female_temp_restrict=df_female_temp.iloc[rd_idx]
                cnt_diff=0
                for idx in range(min_words):
                    input_string_male=df_male_temp['text'].iloc[idx]
                    input_string_female=df_female_temp_restrict['text'].iloc[idx]
                    count_word_male = sum(1 for _ in re.finditer(r'\b%s\b' % re.escape(word), input_string_male.lower()))
                    count_word_female= sum(1 for _ in re.finditer(r'\b%s\b' % re.escape(word), input_string_female.lower()))
                    count_word_diff = count_word_female-count_word_male
                    cnt_diff+=count_word_diff
                word_year_count.append(count_word_diff)
            word_list.append(word)
            year_list.append(year)
        whole_count.append(word_year_count)
    #print("whole_count:",whole_count)
    index = list(range(year_min,year_max))
    data = np.array(whole_count)
    data = data.flatten()
    d={'m <== Δ freq ==> f':data}
    word_array = np.array(word_list)
    wide_df = pd.DataFrame(d)
    wide_df['word']=word_list
    wide_df['year']=year_list
    g=sns.lmplot(x="year", y='m <== Δ freq ==> f', data=wide_df,hue='word',markers=['*', '.','x'],lowess=True);
    g.set(ylim=(-10, 10))
```
